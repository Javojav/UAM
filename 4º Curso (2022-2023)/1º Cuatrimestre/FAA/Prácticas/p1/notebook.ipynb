```python
from Plantillas import Datos, EstrategiaParticionado
from Plantillas.Clasificador import ClasificadorNaiveBayes
from Plantillas.Clasificador import ClasificadorNaiveBayesSk
import numpy as np
import pandas as pd

from sklearn.naive_bayes import GaussianNB as gnb
from sklearn.naive_bayes import CategoricalNB as cnb
from sklearn.naive_bayes import MultinomialNB as mnb
from sklearn import preprocessing
```

```python
def print_table(title, means_and_desvs):
    print(title)
    print("\t|  mean  |  desv  |")
    print("\t -------- --------")
    for tuple in means_and_desvs:
        print("%s\t| %.4f | %.4f |" % tuple)
```

```python
dataPath = './ConjuntoDatos/'

# Carga de datasets
german = Datos.Datos(dataPath + "german.csv")
tictactoe = Datos.Datos(dataPath + 'tic-tac-toe.csv')

# Carga de estrategias
estrategiaVS = EstrategiaParticionado.ValidacionSimple(10, 0.3)
estrategiaVC = EstrategiaParticionado.ValidacionCruzada(0.1)
```

```python
# Carga de clasificadores
nb = ClasificadorNaiveBayes()
multi_nb = ClasificadorNaiveBayesSk(mnb()) 
gauss_nb = ClasificadorNaiveBayesSk(gnb())
categ_nb = ClasificadorNaiveBayesSk(cnb())
```

**Naive Bayes Propio**

```python
# Carga de validaciones german VS
german_errorVS_sinLaplace = nb.validacion(estrategiaVS, german, nb, laplace=False)
german_errorVS_conLaplace = nb.validacion(estrategiaVS, german, nb, laplace=False)

# Carga de validaciones german VC
german_errorVC_sinLaplace = nb.validacion(estrategiaVC, german, nb, laplace=True)
german_errorVC_conLaplace = nb.validacion(estrategiaVC, german, nb, laplace=True)

means_and_desvs = []

#Para validación simple
# Calculo de media y desviacion estandar german sinLaplace
mean_g_vs_sL = np.mean(german_errorVS_sinLaplace)
desv_g_vs_sL = np.std(german_errorVS_sinLaplace)
means_and_desvs.append(('g_vs_sL', mean_g_vs_sL, desv_g_vs_sL))

# Calculo de media y desviacion estandar german conLaplace
mean_g_vs_cL = np.mean(german_errorVS_conLaplace)
desv_g_vs_cL = np.std(german_errorVS_conLaplace)
means_and_desvs.append(('g_vs_cL', mean_g_vs_cL, desv_g_vs_cL))

#Para validación cruzada
# Calculo de media y desviacion estandar german sinLaplace
mean_g_vc_sL = np.mean(german_errorVC_sinLaplace)
desv_g_vc_sL = np.std(german_errorVC_sinLaplace)
means_and_desvs.append(('g_vc_sL', mean_g_vc_sL, desv_g_vc_sL))

# Calculo de media y desviacion estandar german conLaplace
mean_g_vc_cL = np.mean(german_errorVC_conLaplace)
desv_g_vc_cL = np.std(german_errorVC_conLaplace)
means_and_desvs.append(('g_vc_cL', mean_g_vc_cL, desv_g_vc_cL))

# Impresion de tabla con resultados
print_table("Resultados german", means_and_desvs)
```

```python
# Carga de validaciones para tictactoe VS
tictactoe_errorVS_sinLaplace = nb.validacion(estrategiaVS, tictactoe, nb)
tictactoe_errorVS_conLaplace = nb.validacion(estrategiaVS, tictactoe, nb)

# Carga de validaciones para tictactoe VC
tictactoe_errorVC_sinLaplace = nb.validacion(estrategiaVC, tictactoe, nb)
tictactoe_errorVC_conLaplace = nb.validacion(estrategiaVC, tictactoe, nb)

means_and_desvs = []

#Para validación simple
# Calculo de media y desviacion estandar tictactoe sinLaplace
mean_t_vs_sL = np.mean(tictactoe_errorVS_sinLaplace)
desv_t_vs_sL = np.std(tictactoe_errorVS_sinLaplace)
means_and_desvs.append(('t_vs_sL', mean_t_vs_sL, desv_t_vs_sL))

# Calculo de media y desviacion estandar tictactoe conLaplace
mean_t_vs_cL = np.mean(tictactoe_errorVS_conLaplace)
desv_t_vs_cL = np.std(tictactoe_errorVS_conLaplace)
means_and_desvs.append(('t_vs_cL', mean_t_vs_cL, desv_t_vs_cL))

#Para validación cruzada
# Calculo de media y desviacion estandar tictactoe sinLaplace
mean_t_vc_sL = np.mean(tictactoe_errorVC_sinLaplace)
desv_t_vc_sL = np.std(tictactoe_errorVC_sinLaplace)
means_and_desvs.append(('t_vc_sL', mean_t_vc_sL, desv_t_vc_sL))

# Calculo de media y desviacion estandar tictactoe conLaplace
mean_t_vc_cL = np.mean(tictactoe_errorVC_conLaplace)
desv_t_vc_cL = np.std(tictactoe_errorVC_conLaplace)
means_and_desvs.append(('t_vc_cL', mean_t_vc_cL, desv_t_vc_cL))

# Impresion de tabla con resultados
print_table("Resultados tictactoe", means_and_desvs)
```

**Breve análisis de los resultados anteriores. Discutir el efecto Laplace**

Como se observa en las tablas de resultados, el efecto de Laplace afecta de distinta forma para cada dataset aunque también afecta el hecho de usar la validación simple o cruzada y la aleatoriedad de las pruebas.

**One Hot Encoding**

```python
# Carga de dataframe tic-tac-toe
oneHot_tictactoe = pd.read_csv(dataPath + 'tic-tac-toe.csv', dtype={'Class':object})
oneHot_tictactoe = np.array(oneHot_tictactoe)

# Carga del codificador y transformacion de los datos
one_hot = preprocessing.OneHotEncoder(sparse=False, dtype=np.int32)
transform_tictactoe = one_hot.fit_transform(oneHot_tictactoe[:, :-1])

class_transform = []
for value in oneHot_tictactoe[:,-1]:
  if value == 'positive':
    class_transform.append(1)
  else:
    class_transform.append(0)

# Creacion de dataset modificado
oneHot_dataset = Datos.Datos(dataPath + 'tic-tac-toe.csv')
oneHot_dataset.datos = np.concatenate((transform_tictactoe, np.transpose([class_transform])), axis=1)

### Hacemos esta "trampa" para no tener problemas a la hora de llamar a 'validacion'
```

**Multinomial Naive Bayes**

```python
# Carga de validaciones para MultinomialNB
m_oneHot_errorVS = multi_nb.validacion(estrategiaVS, oneHot_dataset, multi_nb)
m_oneHot_errorVC= multi_nb.validacion(estrategiaVC, oneHot_dataset, multi_nb)

means_and_desvs = []

#Para validación simple
m_mean_oneHot_vs= np.mean(m_oneHot_errorVS)
m_desv_oneHot_vs= np.std(m_oneHot_errorVS)
means_and_desvs.append(('m_t_vs', m_mean_oneHot_vs, m_desv_oneHot_vs))

#Para validación cruzada
m_mean_oneHot_vc = np.mean(m_oneHot_errorVC)
m_desv_oneHot_vc = np.std(m_oneHot_errorVC)
means_and_desvs.append(('m_t_vc', m_mean_oneHot_vc, m_desv_oneHot_vc))

# Impresion de tabla con resultados
print_table("Resultados MultinomialNB", means_and_desvs)
```

**Gaussian Naive Bayes**

```python
# Carga de validaciones para GaussianNB
g_oneHot_errorVS = gauss_nb.validacion(estrategiaVS, oneHot_dataset, gauss_nb)
g_oneHot_errorVC= gauss_nb.validacion(estrategiaVC, oneHot_dataset, gauss_nb)

means_and_desvs = []

#Para validación simple
g_mean_oneHot_vs= np.mean(g_oneHot_errorVS)
g_desv_oneHot_vs= np.std(g_oneHot_errorVS)
means_and_desvs.append(('g_t_vs', g_mean_oneHot_vs, g_desv_oneHot_vs))

#Para validación cruzada
g_mean_oneHot_vc = np.mean(g_oneHot_errorVC)
g_desv_oneHot_vc = np.std(g_oneHot_errorVC)
means_and_desvs.append(('g_t_vc', g_mean_oneHot_vc, g_desv_oneHot_vc))

# Impresion de tabla con resultados
print_table("Resultados GaussianNB", means_and_desvs)
```

**Categorical Naive Bayes**

```python
# Carga de validaciones para CategoricalNB
c_oneHot_errorVS = categ_nb.validacion(estrategiaVS, oneHot_dataset, categ_nb)
c_oneHot_errorVC= categ_nb.validacion(estrategiaVC, oneHot_dataset, categ_nb)

means_and_desvs = []

#Para validación simple
c_mean_oneHot_vs= np.mean(c_oneHot_errorVS)
c_desv_oneHot_vs= np.std(c_oneHot_errorVS)
means_and_desvs.append(('c_t_vs', c_mean_oneHot_vs, c_desv_oneHot_vs))

#Para validación cruzada
c_mean_oneHot_vc = np.mean(c_oneHot_errorVC)
c_desv_oneHot_vc = np.std(c_oneHot_errorVC)
means_and_desvs.append(('c_t_vc', c_mean_oneHot_vc, c_desv_oneHot_vc))

# Impresion de tabla con resultados
print_table("Resultados CategoricalNB", means_and_desvs)
```

**Analiza los resultados con los métodos de SKL aplicando OneHotEncoder y compáralos con los obtenidos para el mismo conjunto sin hacer esa transformación**

Los resultados obtenidos informan que OneHotEncoder mejora la efectividad de MultinomialNB, pero empeora a CategoricalNB y un poco a GaussianNB.

**Comparar y analizar los resultados propios con los de SKL**

Para "german.csv", nuestro Naive Bayes es más efectivo que MultinomialNB y prácticamente igual de efectivo que GaussianNB.

Para "tic-tac-toe.csv", nuestro Naive Bayes es igual de efectivo que MultinomialNB, aunque con OneHotEncoder es peor puesto que MultinomialNB mejora; es más o menos parecido a GaussianNB, con y sin OneHotEncoder; y esto último pasa también con CategoricalNB
